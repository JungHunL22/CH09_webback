## 서포트 벡터 머신 (Support Vector Machine)  

서포트 벡터 머신 (Support Vector Machine) 
- 선형이나 비선형 분류, 회귀, 이상치 탐색에도 사용 가능한 다목적 머신러닝 알고리즘
- 데이터가 어느 카테고리에 속할지 판단하는 
- 비확률적 이진 선형 분류 모델을 만들때 사용
- 만들어진 분류 모델은 데이터가 사상된 공간에서 경계로 표현되는데
- 그 중 가장 큰 폭을 가진 경계를 찾는 알고리즘
- 복잡한 분류 문제 또는 작거나 중간 크기의 데이터셋에 적합

**대부분의 분류 모델들은 분류 오차를 최소화 하려는 목적으로 설계되었지만  
SVM은 두 그 사이에 존재하는 마진(간격)을 최대화하여 그룹을 분류하도록 설계  
라지 마진 분류라고도 함** 

- 분류선(Decision Boundary) : 결정 경계
    - 데이터들의 위치에 따라 결정 경계의 위치 달라짐  
- 서포트 벡터(Support Vector) : 분류선과 가장 가까운 데이터  
    - 이 데이터들이 결정 경계를 지지(support)하고 있다고 말할 수 있기 때문에,   
      서포트벡터라고 불림  
- 마진 :  결정 경계와 서포트 벡터 사이의 거리 

서포트 벡터 머신 알고리즘 이해  
- 각 데이터 항목을 다차원의 벡터 공간에 점으로 표시  
- 예로 2차원 공간에 있는 두 개의 그룹을 분류한다고 가정하면  
    - 두 그룹을 매우 잘 구분하는 직선을 찾는 것이 서포트 벡터 머신의 핵심  
    - 데이터를 선형 또는 비선형으로 분리하는 최적의 경계를 찾는 알고리즘 
    - 최적의 결정 경계는 마진을 최대화 하는 것

#### 서포트 벡터 머신의 동작 원리   

**선형 SVM : 데이터를 선형으로 분리하는 최적의 선형 결정 경계를 찾는 알고리즘**

두 단계에 따라 클러스터를 최대한 잘 구분하도록 작업 수행     
(1) 첫 번째 : 두 개의 그룹을 분리하는 직선을 찾는  
(2) 두 번째 : 균형있게 나눈다   

(1) 첫 번째 : 두 개의 그룹을 분리하는 직선을 찾는다  
    - 이러한 직선을 통한 이진 분류 기법은   
    - N개의 그룹으로 분류화하는 문제를 쉽게 처리할 수 있음      

(2) 두 번째 : 균형있게 나눈다 (마진 최대화)    
- 한쪽에 치우치지 않게 양쪽 그룹 모두 균형 있게 나누려면    
- 두 그룹을 식별하는 직선을 구하되    
- 직선식을 사이에 두고 가능하면 최대한 멀리 떨어지도록 점선을 구함  
- 직선식에 가장 멀리 떨어진 이 거리를 '최대 마진'이라고 함  
- 점선에 속한 점이 직선을 침범하지 않도록 만드는 안전지대 역할을 하고,     
- 버팀대 또는 지지대 역할을 한다고 해서 서포트 벡터라고 부름

서포트 벡터 머신의 핵심
- 훈련이 진행됨에 따라 분류 기준은 점점 더 많은 데이터 샘플을 보게 되므로,   
- 최적의 경계가 어디에 있는지를 찾게 됨 
- 이 직선 사이의 거리인 마진이 최대가 되도록 직선을 조정하여 
- 두 그룹을 최대한 분리시키는 것

서포트 벡터 머신의 장점
- 속도가 빠르다
    - 대부분의 머신러닝 지도학습 알고리즘들이
        - 학습 데이터를 모두 사용하여 모델을 학습하는 반면
    - **SVM은 결정 경계를 정의하는 서포트 벡터만 잘 골라내면** 
        - 나머지 중요하지 않은 수 많은 데이터 포인트들을 무시할 수 있기 때문에
        - 속도가 매우 빠름

**SVM에서 결정 경계를 구하는 건 상당히 복잡한 최적화 문제**

- 하드 마진(hard margin)
    - 모든 학습 샘플이 결정 경계 바깥쪽에 올바르게 분류되도록 하는 마진 분류   

        - 훈련 세트가 선형적으로 구분되는 경우에만 가능
            - 아웃라이어를 허용하지 않고 기준을 까다롭게 설정
        - 서포트 벡터와 결정 경계 사이의 거리가 좁음
            - 즉, 마진이 작아짐
        - 이상치 데이터까지 포함시켜 결정 경계를 정하면 
            - 오버피닝(과적합) 발생 가능

      
- 소프트 마진(soft margin)
    - 마진 위반(margin violation) 사례의 발생 정도를 조절하면서
        - 결정 경계의 폭을 최대로 넓게 유지하는 마진 분류
    - 마진 위반: 학습 샘플이 결정 경계를 넘어 해당 클래스 반대편에 위치하는 샘플
        - 마진 안에 아웃라이어 포함시켜 결정 경계를 느슨하게 설정
    - 소프트 벡터와 결정 경계 사이의 거리가 멀어짐
        - 마진이 커짐
        - 이경우 언더피팅(과소적합) 발생 가능

- 사이킷런의 SVM 클래스 최적화하기
    - 하드마진과 소프트 마진을 설정하기 위해 C(cost) 파라미터 사용
    - C : 규제 강도
        - 값이 클수록 하드마진
        - 값이 작을 수록 소프트 마진
    - 적절한 C값을 찾는 것이 중요


서포트 벡터 머신 유형
- 선형 SVM
- 비선형 SVM
    - 커널 기법 사용
    - (입력 자료의) 다차원 공간상으로의 맵핑(mapping) 기법을 사용하여 
    - 비선형분류도 효율적으로 수행

### 비선형 SVM

- 선형적으로 도저히 구별하기 어려운 특징은??? 
    - 오른쪽 그림처럼 비선형 방법을 사용


- 구별이 가능한 방향으로 mapping을 시키면 
    - 구별이 불가능한 특징들도 새로운 공간에서 구별이 가능


- 커널 함수
    - 아래 그림처럼 평면에서는 불가능한 구분을 고차원으로 변경 후 분류할 때 사용하는 함수


커널 기법
- 주어진 데이터를 고차원 특징 공간으로 사상해 주는 것
- 고차원 공간에 사상되고 나면 
- 원래의 차원에서는 보이지 않던 선형 분류 방법이 보임

SVM에서 데이터를 고차원 공간에 매핑하는 방법은 크게 2가지  
1. 다항식 커널 SVM  
    - 개체간의 고차원 상의 관계를 계산하여 소프트 마진 분류기를 찾아내는 방식
    - 원래 특성의 가능한 조합을 지정된 차수까지 모두 계산 
2. RBF(radical basis function) 커널 SVM 
    - 방사 기저 함수
    - 가우시안 분포와 유사해서 가우시안 커널로도 불림  
    - 유사성 또는 서로 얼마나 가까운지를 계산
    - (다항식 커널의 경우 일부 지정된 차수 까지만 계산하지만    
       RBF는 무한하게 확장)  
    - 모든 차수의 모든 다항식을 고려
    - 선형 SVM으로는 분류하기 어려운 경우 가장 널리 활용

```python
def plot_iris(X, y, model, title, xmin=-2.5, xmax=2.5, ymin=-2.5, ymax=2.5):
    XX, YY = np.meshgrid(np.arange(xmin, xmax, (xmax-xmin)/1000),
                         np.arange(ymin, ymax, (ymax-ymin)/1000))
    ZZ = np.reshape(model.predict(np.array([XX.ravel(), YY.ravel()]).T), XX.shape) 
    plt.contourf(XX, YY, ZZ, cmap=mpl.cm.Paired_r, alpha=0.5) # 등고선도
    plt.scatter(X[y == 0, 0], X[y == 0, 1], c='r', marker='^', label='0', s=100)
    plt.scatter(X[y == 1, 0], X[y == 1, 1], c='g', marker='o', label='1', s=100)
    plt.scatter(X[y == 2, 0], X[y == 2, 1], c='b', marker='s', label='2', s=100)
    plt.xlim(xmin, xmax)
    plt.ylim(ymin, ymax)
    plt.xlabel("꽃잎의 길이")
    plt.ylabel("꽃잎의 폭")
    plt.title(title)
    
plt.figure(figsize=(8, 12))
plt.subplot(311)
plot_iris(X_test_std, y_test, model1, "선형 SVC")
plt.subplot(312)
plot_iris(X_test_std, y_test, model2, "다항커널 SVC")
plt.subplot(313)
plot_iris(X_test_std, y_test, model3, "RBF커널 SVM")
plt.tight_layout()
plt.show()
```
