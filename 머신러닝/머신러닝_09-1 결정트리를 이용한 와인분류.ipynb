## 결정트리 예제 1

```python
### 로지스틱 회귀로 와인 분류
```

```python
import pandas as pd

wine = pd.read_csv('https://bit.ly/wine-date')
```

```python
wine.head()
```

```python
wine.info()
```

```python
wine.describe()
```

```python
data=wine[['alcohol','sugar','pH']].to_numpy()
target=wine['class'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    data, target, test_size=0.2, random_state=42)
```

```python
# 로지스틱 회귀 알고리즘이고 sugar의 값의 범위가 다른 특성에 비해 차이가 너무 크기 때문에 표준화 작업 진행
# 표준화 변환 진행
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(train_input) # 학습데이터 이용해 변환기준을 만든 뒤 테스트데이터는 학습데이터 기준으로 표준화 진행

train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(train_scaled, train_target)

print(lr.score(train_scaled, train_target))
print(lr.score(test_scaled, test_target))
```

### 모델 설명

```python
print(lr.coef_,lr.intercept_)
```

- 이 모델은
- 알코올 * 0.51270274 + 당도 * 1.6733911 + pH * -0.68767781 + 1.81777902
    - 0보다 크면 화이트 와인
    - 작으면 레드와인으로 분류
- 정확도는 77%인 모델

### 결정트리 모델링

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(train_scaled, train_target)

print(dt.score(train_scaled, train_target))
print(dt.score(test_scaled, test_target))
# 하이퍼 파라미터 튜닝하지 않아서 과적합 발생(Train데이터에 과적합돼서 다른 데이터에는 정확도가 낮아짐)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(10,7))
plot_tree(dt, max_depth=1, filled=True, feature_names=['alcohol', 'sugar', 'pH'])
plt.show()
```

### 가지치기 : max_depth 하이퍼 파라미터를 제한

```python
dt = DecisionTreeClassifier(max_depth=3, random_state=42)
dt.fit(train_input, train_target)

print(dt.score(train_input, train_target))
print(dt.score(test_input, test_target))
```

```python
plt.figure(figsize=(20,15))
plot_tree(dt, filled=True, feature_names=['alcohol', 'sugar', 'pH'])
plt.show()
```
