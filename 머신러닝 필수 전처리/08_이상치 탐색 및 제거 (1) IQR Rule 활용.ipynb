```python
# 특성 분리
X=df.drop(['Type'],axis=1)
Y=df['Type']
```

```python
# 학습 데이터와 평가 데이터 분리
from sklearn.model_selection import train_test_split
Train_X, Test_X, Train_Y, Test_Y = train_test_split(X, Y)
Train_X.shape
```

```python
import numpy as np
```

```python
def IQR_rule(var_list): # 한개의 특성 배열을 파라미터로 받음
    # IQR 계산
    Q1 = np.quantile(var_list,0.25)
    Q3 = np.quantile(var_list,0.75)
    IQR = Q3-Q1

    # IQR Rule의 위배 결과, bool list를 생성 후에 반환
    # 이상치의 조건 : Q3 + IQR*1.5 초과 / Q1 - IQR*1.5 미만이면 이상치
    # 결과값이 True면 정상데이터, 결과값이 False면 이상치
    not_outlier_condition = (Q3 + 1.5 * IQR > var_list) & (Q1 - 1.5 * IQR < var_list)
    return not_outlier_condition
    
```

```python
# apply() 함수를 이용해서 모든 컬럼에 IQR_rule 함수를 적용
conditions=Train_X.apply(IQR_rule)
conditions
```

```python
# 레코드에서 관측치 중 하나라도 이상치면 해당 레코드를 제거
conditions.sum(axis=1)==8 # 이 결과가 True면 레코드 사용 False면 레코드 제거
```

```python
total_condition=conditions.sum(axis=1)==8
# 하나라도 IQR 규칙을 위반하는 요소를 갖는 레코드를 제거하기 위한 규칙
```

```python
Train_X=Train_X.loc[total_condition]
```
