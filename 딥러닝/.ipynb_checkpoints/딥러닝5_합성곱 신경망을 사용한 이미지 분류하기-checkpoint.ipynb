## 합성곱 신경망을 사용한 이미지 분류

### 패션 MNIST 데이터 불러오기

```python
from tensorflow import keras
from sklearn.model_selection import train_test_split
```

```python
(train_input,train_target),(test_input,test_target)=keras.datasets.fashion_mnist.load_data()

train_scaled = train_input.reshape(-1,28,28,1)/255.0
# 깊이정보를 전달했으므로 4차원 data로 구성됨
# 이미지 1장이 3차원 데이터로 구성*1까지 계산해서 4차원으로 구성
# 이미지 픽셀값의 범위를 줄이고 3차원 이미지로 생성

train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_scaled, train_target, test_size=0.2, random_state=42)
```

```python
train_scaled.shape
# 28*28*1이 이미지 한장의 차원
```

## 합성곱 신경망 만들기

```python
model=keras.Sequential()
```

### 첫번째 합성곱 - 풀링 층 추가

```python
model.add(keras.layers.Conv2D(32,kernel_size=(3,3),activation='relu',padding='same',
                             input_shape=(28,28,1)))
# 32개의 필터가 특성맵 32개를 생성
```

```python
# 특성맵크기를 줄이기 위해(다음층으로 전달되는 값의 크기를 줄이기 위해)
# 풀링( 특성맵의 크기를 1/2로 줄임)
model.add(keras.layers.MaxPooling2D(2))
```

### 두번째 합성곱 - 풀링 층 추가

```python
model.add(keras.layers.Conv2D(64, kernel_size=(3,3), activation='relu',
                              padding='same'))
# 2*2 픽셀의 값 중 가장 큰 값만 사용하는 풀링방식
model.add(keras.layers.MaxPooling2D(2))
```

### 세번째 : 3차원 특성맵을 일렬로 펼침

- 마지막에 10개의 뉴런을 가진 출력(밀집)층에서 확률을 계산해야 하기 떄문에 일렬로 펼쳐야 한다

### 네번째 : 은닉층 추가 - Dropout(과대적합 방지)

```python
model.add(keras.layers.Flatten()) # 평면화
model.add(keras.layers.Dense(100,activation='relu')) # 은닉층 100개
model.add(keras.layers.Dropout(0.4)) # 40%씩 드롭아웃
```

### 다섯번째 : 최종 출력층 추가

```python
model.add(keras.layers.Dense(10,activation='softmax')) # 출력층 10개
```

```python
model.summary() # 너비*높이*깊이?
```

- 합성곱 층과 풀링층 효과가 잘 나타나 있음
    - 첫번째 합성곱 층을 통과하면
        - 특성 맵의 길이는 32가 됨
    - 두번째 합성곱에서 맵의 깊이는 64로 늘어남
    - but, 특성맵의 가로세로 크기는 첫번째 풀링층에서 절반으로 줄어들고 두번째 풀링층에서 절반으로 줄어들어 (7,7,64)가 됨
- 파라미터 갯수
    - 첫번째 합성곱
        - 32개의 필터, 크기가 (3,3), 깊이가 1, 필터마다 절편이므로 9x9x1x32+32=320

## 모델 컴파일과 훈련

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics='accuracy')
checkpoint_cb=keras.callbacks.ModelCheckpoint('best-cnn-model.h5')
early_stopping_cb=keras.callbacks.EarlyStopping(patience=2,
                                               restore_best_weights=True)
# 종료된 시점에서 2단계 뒤의 가중치를 저장하게 됨
history=model.fit(train_scaled,train_target,epochs=20,
                 validation_data=(val_scaled,val_target),
                 callbacks=[checkpoint_cb,early_stopping_cb])
```

### 결과 시각화

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train', 'val'])
plt.show()
```

- 7번째(6번)에포크에서 손실 최저를 찍고 다시 증가하고있는 것을 볼 수 있음
    - 최적 에포크는 7번

```python
model.evaluate(val_scaled,val_target)
```

- 92퍼센트 정확도가 나옴
