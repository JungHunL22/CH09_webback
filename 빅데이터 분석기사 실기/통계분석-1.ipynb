# Scipy를 활용한 상관분석
Scipy패키지의 stats 서브패키지를 호출한 후 stats에서 제공하는 pearsonr함수를 사용

```python
import pandas as pd
from sklearn.datasets import load_diabetes
```

```python
dia=load_diabetes()
```

```python
dia
```

```python
dia_data=pd.DataFrame(dia['data'],columns=dia['feature_names'])
```

```python
dia_data['target']=dia['target']
```

```python
dia_data
```

## 피어슨 상관계수(연속형 변수 간의 상관분석)

나이와 타겟값의 상관계수 0.18로 낮지만, p_value가 0.05보다 매우작아 두 컬럼 간의 차이가 유의하다고 할 수 있음.

```python
from scipy.stats import pearsonr
pearsonr(dia_data['age'],dia_data['target'])
```

타겟 컬럼과 가장 상관관계가 높은 컬럼은 bmi, 가장 상관관계가 낮은 컬럼은 sex로 확인

```python
dia_data.corr()
```

## 스피어만 상관계수(주로 범주형이 포함된 변수 간의 상관분석)

성별과 타겟값의 상관계수는 0.03으로 낮으며, p_value가 0.4이므로 두 컬럼 간의 차이는 유의하지 않다고 할 수 있음

```python
from scipy.stats import spearmanr
spearmanr(dia_data['sex'],dia_data['target'])
```

```python
dia_data.corr(method='spearman')
```
