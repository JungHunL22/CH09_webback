```python
# 이미지 다운로드
import urllib.request as req
url="http://uta.pw/shodou/img/28/214.png"
req.urlretrieve(url,'test.png')
```

### OpenCV로 읽어 들이기 
- imread()
    - 색상공간 BGR 형태로 이미지를 읽는다
    - 대표적인 색상공간
        - Gray/HSV(색상,채도,명도)/RGB/LAB(Lightness,Green to Magenta, Blue to Yellow)
        - 다른 기능에 이미지를 사용하기 위해서는 BGR => RGB로 변경해야 함


```python
img=cv2.imread('test.png')
print(img.shape)
# width,height, 한 픽셀당 색상 데이터 수
print(img)
```

```python
# imread()는 에러를 발생시키지 않는다
# 없는 파일을 읽었을 때 None값이 나오므로 출력해서 확인해야함
imgtest=cv2.imread('존재하지않는 파일.png')
print(imgtest)
```
