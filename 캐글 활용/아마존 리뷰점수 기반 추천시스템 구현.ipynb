```python
import pandas as pd
import numpy as np
import random
```

```python
review=pd.read_csv('data/amazon_ratings.csv',names = ['User ID', 'Product ID', 'Rating','Time Stamp']).sample(n=100000)
review.head()
```

```python
review.shape
```

```python
matrix=rv.pivot_table('Rating',index=['User ID'],columns=['Product ID']).fillna(0)
```

```python
matrix
```

```python
matrix.shape
```

```python
mat=matrix.T
```

```python
mat
```

```python
mat.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cos=cosine_similarity(mat,mat)
```

```python
cos.shape
```

```python
df=pd.DataFrame(cos,index=mat.index,columns=mat.index)
```

영화 간 유사도 행렬

```python
df.shape
```

```python
rating=matrix.values.dot(df.values)/np.array([np.abs(df.values).sum(axis=1)])
```

```python
rating.shape
```

```python
rat=pd.DataFrame(rating,index=matrix.index,columns=matrix.columns)
```

```python
rat
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
def mse(pred,actual):
    pred = pred[actual.nonzero()].flatten()
    actual = actual[actual.nonzero()].flatten()
    return mean_squared_error(actual,pred),r2_score(actual,pred)
```
