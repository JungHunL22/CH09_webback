# Movielens 데이터 추천시스템 구현(Matrix Factorization)

```python
import pandas as pd
import numpy as np
```

```python
movie=pd.read_csv('F:/Data/movies.csv')
rating=pd.read_csv('F:/Data/ratings.csv')
```

```python
movie
```

```python
rating
```

```python
matrix=rating.pivot_table('rating',index=['userId'],columns=['movieId']).fillna(0)
```

```python
matrix
```

# 사용자-영화별 평균평점

```python
urat_mean=matrix.mean(axis=1).values.reshape(-1,1)
```

```python
up_mat=matrix-urat_mean
```

```python
up_mat
```

```python
from scipy.sparse.linalg import svds
from scipy.linalg import svd
```

```python
u,s,v=svds(up_mat,k=10)
```

```python
print(u.shape,s.shape,v.shape)
```

```python
s_diag=np.diag(s)
```

```python
mat_rating=pd.DataFrame(u.dot(s_diag).dot(v),index=up_mat.index,columns=up_mat.columns)
```

```python
preds=mat_rating+up_mat
```

```python
preds
```

```python
sort_r=preds.loc[300].sort_values(ascending=False)
sort_r
```

```python
user_dt=rating[rating.userId==300]
user_dt
```

```python
user_his=user_dt.merge(movie,on='movieId').sort_values('rating',ascending=False)
user_his
```

```python
re=movie[~movie['movieId'].isin(user_his['movieId'])]
re
```

```python
re_res=re.merge(pd.DataFrame(sort_r).reset_index(),on='movieId')
```

```python
re_res
```

# 입력함수

```python
def recomend(preds,user_id,rating,movie,num=10):
    # 사용자에 따른 데이터 정렬(점수기준 내림차순)
    sort_user_pred=preds.loc[user_id].sort_values(ascending=False)
    
    # userid 정보
    user_dt=rating[rating.userId==user_id]
#     print(user_dt)
    user_his=user_dt.merge(movie,on='movieId').sort_values('rating',ascending=False)
#     print(user_his)
    recommend=movie[~movie['movieId'].isin(user_his['movieId'])]
#     print(recommend)
    recommend_res=recommend.merge(pd.DataFrame(sort_user_pred).reset_index(),on='movieId')
#     print(recommend_res)
    
#     recommend_res=recommend.rename(columns={user_id:'Predict'}).sort_values('Predict')
    recommend_res=recommend_res.rename(columns={user_id:'Predict'}).sort_values('Predict',ascending=False)[:num]
#     print(recommend_res)
    
    return user_his,recommend_res
```

```python
history,res=recomend(preds,200,rating,movie)
```
