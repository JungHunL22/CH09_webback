# 사용자-상품별 평균평점

```python
urat_mean=matrix.mean(axis=1).values.reshape(-1,1)
```

```python
up_mat=matrix-urat_mean
```

```python
up_mat
```

```python
from scipy.sparse.linalg import svds
from scipy.linalg import svd
```

```python
u,s,v=svds(up_mat,k=10)
```

```python
print(u.shape,s.shape,v.shape)
```

```python
s_diag=np.diag(s)
```

```python
mat_rating=pd.DataFrame(u.dot(s_diag).dot(v),index=up_mat.index,columns=up_mat.columns)
```

```python
preds=mat_rating+up_mat
```

# 입력함수

```python
def recomend(preds,user_id,rating,product,num=10):
    # 사용자에 따른 데이터 정렬(점수기준 내림차순)
    sort_user_pred=preds.loc[user_id].sort_values(ascending=False)
    
    # userid 정보
    user_dt=rating[rating.User_ID==user_id]
#     print(user_dt)
    user_his=user_dt.merge(product,on='Product_ID').sort_values('Rating',ascending=False)
#     print(user_his)
    recommend=product[~product['Product_ID'].isin(user_his['Product_ID'])]
#     print(recommend)
    recommend_res=recommend.merge(pd.DataFrame(sort_user_pred).reset_index(),on='Product_ID')
#     print(recommend_res)
    
#     recommend_res=recommend.rename(columns={user_id:'Predict'}).sort_values('Predict')
    recommend_res=recommend_res.rename(columns={user_id:'Predict'}).sort_values('Predict',ascending=False)
#     print(recommend_res)
    
    return user_his,recommend_res
```

```python
history,res=recomend(preds,'AZX2B0WYJECLO',review,product)
```
